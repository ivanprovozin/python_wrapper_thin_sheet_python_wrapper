```python
from pathlib import Path
from concurrent.futures import ThreadPoolExecutor 
from wsheet20 import Wsheet20, Sheet, ColeCole, Source

def create_run_for_frequency(freq_hz, output_dir):
    run = Wsheet20(
        sheets=[
            Sheet(
                na=20, nb=20,
                center=(500., 0., -200.),
                strike_length=200., dip_length=200.,
                strike=90., dip=90., thickness=0.1,
                resistivity=ColeCole(low=1e-5, high=1e-5, tau=0., alpha=0.5),
                permittivity=ColeCole(low=2., high=2., tau=0., alpha=0.5),
            )
        ],
        host_resistivity  = ColeCole(low=10., high=10., tau=0., alpha=0.5),
        host_permittivity = ColeCole(low=10., high=10., tau=0., alpha=0.5),
        frequency         = freq_hz,
        source            = Source(type="magnetic_dipole", direction="z", angle=0.),
        n_tx              = 1,
        tx_start          = (0., 0., 0.),
        tx_increment      = (0., 0., 0.),
        n_rx              = 50,
        rx_start          = (0., 0., 0.75),
        rx_increment      = (5., 0., 0.),
        green_mode        = 0,
        output_scatter    = True,
        output_efield     = True,
        output_hfield     = True,
        normalize         = False,
    )
    
    # Set executable and working directory
    run.executable = Path('/home/vanza334/projects/def-sgkang09/vanza334/python_wrapper') / "a.out"
    
    # Create output directory per frequency
    freq_dir = output_dir / f"freq_{int(freq_hz)}"
    freq_dir.mkdir(exist_ok=True)
    run.work_dir = freq_dir
    
    print(f"Starting run for frequency: {freq_hz} Hz")
    run.run()
    print(f"Finished run for frequency: {freq_hz} Hz")
    return freq_hz, freq_dir

# Define frequencies to run
frequencies = [5e3, 10e3, 20e3, 40e3, 80e3]  # example frequencies in Hz

# Base output directory
output_dir = Path('/home/vanza334/projects/def-sgkang09/vanza334/python_wrapper/output')
output_dir.mkdir(exist_ok=True)

# Run simulations in parallel (match max_workers to CPUs using)
with ThreadPoolExecutor(max_workers=5) as executor:
    futures = [executor.submit(create_run_for_frequency, f, output_dir) for f in frequencies]
    for future in futures:
        freq, outdir = future.result()
        print(f"Completed frequency {freq} Hz, results in {outdir}")
```
